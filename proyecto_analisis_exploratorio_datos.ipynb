Antes que nada se debe formatear la columna **Periodo** que contiene solo los años,y su tipo de dato es object, como planeo hacer comparaciones de tiempo para poder ordenar o realizar alguna grafica de serie temporal que se relacione a las otras variables, para ello voy a convertir esos rangos de años a fechas fijas y a convertirlo a un tipo de dato datatime, asignando un día y mes fijos, como el primero de enero de cada año, para mantener la consistencia.

Tambien pude notar que la columna **Periodo**  contiene años en algunas filas separados por un guion entonces para procesar la columna **Periodo** , utilizare una función que toma cada valor de la columna y verifica si contiene un rango de años que esta indicado por la presencia de un guion (**-**), y en ese caso, debe extraer y devolver el primer año del rango. Si no hay un rango, simplemente devuelve el valor del periodo tal como está.

```python
# Función para extraer el primer año del periodo
def extract_first_year(Periodo):
    # Si el periodo contiene un rango de años (e.g., "2007 - 2008")
    if '-' in str(Periodo):
        return str(Periodo).split(' - ')[0]
    return str(Periodo)

# Aplicar la función 'extract_first_year' a la columna 'Periodo' del DataFrame 'df'
df_mercado_laboral['Periodo'] = df_mercado_laboral['Periodo'].apply(extract_first_year)

# Convertir la columna 'Periodo' actualizada a datetime especificando el formato
df_mercado_laboral['Periodo'] = pd.to_datetime(df_mercado_laboral['Periodo'] + '-01-01', format='%Y-%m-%d')

# Verificar la conversión
print(df_mercado_laboral['Periodo'].head(15))
print(df_mercado_laboral['Periodo'].dtype)
```

Para los datos faltantes, se utilizara el promedio trimestral o anual para completar los datos faltantes. Este método es menos susceptible a errores y proporciona una aproximación razonable.

```python
# Ver los nombres exactos de las columnas
print(df_mercado_laboral.columns)
```

```python
# Establecer un índice multi-nivel para la imputacion
df_mercado_laboral.set_index(['Periodo', 'Trimestre', 'Trimestre móvil cronológico'], inplace=True)
```

Imputación con el Promedio Trimestral:
Se agrupan los datos por la columna **Periodo** y luego reemplaza los valores faltantes (NaN) en cada grupo con el promedio (media) de ese grupo.

Esto significa que para cada **Periodo**, todos los valores faltantes se reemplazarán con el promedio de los valores no faltantes del mismo **Periodo**.

Después de la primera imputación, se agrupan nuevamente los datos por **Trimestre móvil cronológico** y realiza la misma operación de reemplazo de valores faltantes con el promedio del grupo.

```python
# Imputar los valores faltantes con el promedio trimestral o anual
df_mercado_laboral = df_mercado_laboral.groupby(['Periodo']).transform(lambda x: x.fillna(x.mean()))
df_mercado_laboral = df_mercado_laboral.groupby(['Trimestre móvil cronológico']).transform(lambda x: x.fillna(x.mean()))
```

# Cargamos la data preprocesada

```python
# Carga el conjunto de datos
import pandas as pd
df_mercado_limpio = pd.read_excel('conjunto-mercado-laboral-limpio1.xlsx')
```

```python
# Mostrar las primeras filas para entender la estructura
print(df_mercado_limpio.head(12))
```

```python
valores_faltantes = df_mercado_limpio.isnull().sum()
print(valores_faltantes)
```

Al verificar la columna periodo tiene valores nulos ya que para hacer la imputacion se eliminaron los duplicados de esta columna para generar un indice, para el tratamiento de estos valores faltantes se creara primero un diccionario para convertir los valores iniciales de la columna trimestre en numeros

```python
# Función para obtener el mes inicial de un trimestre
def obtener_mes_inicio(trimestre):
    mes_inicial = trimestre.split(' ')[0]
    return meses_dict[mes_inicial]
```

```python
# Diccionario para convertir las iniciales de los meses a números de mes
meses_dict = {
    'ene': 1, 'feb': 2, 'mar': 3, 'abr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'ago': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dic': 12
}
```

Se utiliza un bucle for para recorrer sobre cada fila del DataFrame con la función iterrows() devuelve el índice de la fila (indice), y dentro del bucle, se verifica si el valor en la columna **‘Periodo’** de la fila actual es un valor nulo (NaN) usando la función pd.isna().

Si el valor en la columna **‘Trimestre móvil cronológico’** es igual a 1, significa que es el primer trimestre del año, y se debe establecer el año base. Se calcula el año actual sumando a 2007 la división entera del índice de la fila más uno entre 12.

Aqui es donde se utiliza la función **obtener_mes_inicio()** con el valor de la columna **‘Trimestre’** de la fila actual. Esta función devuelve el mes de inicio del trimestre.

Se actualiza la columna **‘Periodo’** de la fila actual con una fecha construida usando el año actual, el mes de inicio y el primer día del mes. La función pd.to_datetime() convierte el nuevo valor a un tipo de dato datatime para realizar series de tiempo.

Si el valor en la columna **‘Trimestre móvil cronológico’** es 12, significa que es el último trimestre del año, y se incrementa año_actual en uno para el siguiente ciclo de trimestres.

```python
# Rellenar la columna 'Periodo'
año_actual = 2007

for indice, fila in df_mercado_limpio.iterrows():
    if pd.isna(fila['Periodo']):
        # Si es el primer trimestre móvil cronológico, establece el año base correcto
        if fila['Trimestre móvil cronológico'] == 1:
            # Calcula el año basado en el índice de la fila y el trimestre móvil cronológico
            año_actual = 2007 + ((indice + 1) // 12)
        mes_inicio = obtener_mes_inicio(fila['Trimestre'])
        df_mercado_limpio.at[indice, 'Periodo'] = pd.to_datetime(f"{año_actual}-{mes_inicio:02d}-01")
        
        # Incrementar el año después de cada conjunto de 12 trimestres móviles cronológicos
        if fila['Trimestre móvil cronológico'] == 12:
            año_actual += 1
```

# ¿Cómo se distribuye la fuerza laboral entre ocupados asalariados, no asalariados e informales?

```python
columns_of_interest = ['Ocupados  asalariados', 'Ocupados no asalariados', 'Total informales DANE ']

# Filtrar y agrupar los datos según las columnas de interés
df_filtered = df_mercado_limpio[columns_of_interest]

# Calcular la suma de cada categoría
category_sums = df_filtered.sum()

# Calcular el porcentaje de cada columna de interes
category_percentage = (category_sums / category_sums.sum()) * 100

# Crear una gráfica de pastel para visualizar la distribución
plt.figure(figsize=(10, 7))
plt.pie(category_percentage, labels=category_percentage.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de la Fuerza Laboral entre Asalariados, No Asalariados e Informales')
plt.axis('equal')  # Para asegurar que el gráfico sea un círculo
plt.show()
```

# Cuartiles e IQR para la tasa de ocupación:

* Outliers:
Los puntos rojos representan los outliers, que son valores atípicos que se encuentran fuera del rango esperado (Q1 - 1.5 * IQR y Q3 + 1.5 * IQR).

* Q1 (Primer Cuartil): Representa el valor por debajo del cual se encuentra el 25% de los datos.
  
* Q3 (Tercer Cuartil): Representa el valor por debajo del cual se encuentra el 75% de los datos.
  
* IQR (Rango Intercuartílico): Es la diferencia entre Q3 y Q1 y mide la dispersión de la mitad central de los datos.
  
* Mediana:
La línea central dentro del cuadro azul representa la mediana de la tasa de ocupación, que es el valor central de los datos.

* Bigotes:
Los bigotes se extienden desde los cuartiles hasta los valores máximos y mínimos que no son considerados outliers.

* Distribución:
La distribución de la tasa de ocupación se visualiza claramente, mostrando la concentración de datos y la presencia de valores atípicos.

```python
# Calcular los cuartiles y el rango intercuartílico (IQR) para la tasa de ocupación
Q1 = df_mercado_limpio['Tasa de ocupacion'].quantile(0.25)
Q3 = df_mercado_limpio['Tasa de ocupacion'].quantile(0.75)
IQR = Q3 - Q1

# Identificar los valores atípicos (outliers) basados en IQR
outliers = df_mercado_limpio[(df_mercado_limpio['Tasa de ocupacion'] < Q1 - 1.5 * IQR) | 
                              (df_mercado_limpio['Tasa de ocupacion'] > Q3 + 1.5 * IQR)]

# Crear el gráfico de caja para visualizar la distribución y los outliers
plt.figure(figsize=(10, 6))
plt.boxplot(df_mercado_limpio['Tasa de ocupacion'], vert=False, patch_artist=True)

# Resaltar los outliers en el gráfico
plt.scatter(outliers['Tasa de ocupacion'], [1]*len(outliers), color='red', label='Outliers')

# Añadir título y etiquetas para los ejes
plt.title('Detección de Outliers en la Tasa de Ocupación')
plt.xlabel('Tasa de Ocupación (%)')
plt.yticks([1], ['Tasa de Ocupación'])

# Añadir leyenda y rejilla para facilitar la lectura
plt.legend()
plt.grid(True)

# Guardar el gráfico en un archivo
plt.savefig("Detección de Outliers en la Tasa de Ocupación.png", bbox_inches='tight')

# Mostrar el gráfico resultante
plt.show()
```

Se utiliza el Z-Score para identificar los outliers.

Los outliers se definen como valores que tienen un Z-score mayor a 3 o menor a -3, lo que indica que están a más de 3 desviaciones estándar de la media.


```python
from scipy.stats import zscore

# Calcular el Z-score para la 'Tasa de ocupacion'
df_mercado_limpio['z_score_ocupacion'] = zscore(df_mercado_limpio['Tasa de ocupacion'])

# Identificar los outliers utilizando el criterio de |Z-score| > 3
outliers_z = df_mercado_limpio[df_mercado_limpio['z_score_ocupacion'].abs() > 3]

# Imprimir los outliers detectados
print("Outliers detectados:\n", outliers_z[['Tasa de ocupacion', 'z_score_ocupacion']])

# Crear la gráfica de caja utilizando Seaborn para una mejor visualización
plt.figure(figsize=(12, 6))
sns.boxplot(x=df_mercado_limpio['Tasa de ocupacion'], orient='h', color='skyblue')

# Utilizar Seaborn para graficar los outliers y controlar el tamaño de los puntos
sns.scatterplot(x=outliers_z['Tasa de ocupacion'], y=[0] * len(outliers_z), color='red', s=100, label='Outliers')

# Añadir título y etiquetas para los ejes
plt.title('Detección de Outliers en la Tasa de Ocupación utilizando Z-Score')
plt.xlabel('Tasa de Ocupación (%)')

# Mostrar la leyenda
plt.legend()

# Guardar el gráfico en un archivo
plt.savefig("Detección de Outliers en la Tasa de Ocupación utilizando Z-Score.png", bbox_inches='tight')

# Mostrar el gráfico resultante
plt.show()
```

## Comparación
* Método de Detección:
  
El primer gráfico utiliza el IQR, que es una medida basada en los cuartiles y es menos sensible a la distribución de los datos.

El segundo gráfico utiliza el Z-score, que es una medida basada en la desviación estándar y es más sensible a la distribución normal de los datos.

En el primer gráfico, los outliers se identifican en función de su posición relativa a los cuartiles y en el segundo gráfico, los outliers se identifican en función de su desviación respecto a la media.

Aunque ambos gráficos utilizan puntos rojos para marcar los outliers, el contexto de identificación es diferente.

El primer gráfico puede ser más adecuado para datos que no siguen una distribución normal, mientras que el segundo gráfico es más adecuado para datos que siguen una distribución normal.

# Z-score modificado
Los outliers se definen como valores que tienen un Z-score modificado extremo, lo que puede ser más robusto frente a distribuciones no normales.

```python
import numpy as np

# Función para calcular el Z-score modificado
def modified_z_score(series):
    median_y = np.median(series)
    mad_y = np.median(np.abs(series - median_y))
    modified_z_scores = 0.6745 * (series - median_y) / mad_y
    return np.abs(modified_z_scores)

# Calcular el Z-score modificado para la 'Tasa de ocupacion'
df_mercado_limpio['modified_z_score_ocupacion'] = modified_z_score(df_mercado_limpio['Tasa de ocupacion'])

# Identificar los outliers utilizando el criterio de |Z-score| > 3.5
outliers_modified_z = df_mercado_limpio[df_mercado_limpio['modified_z_score_ocupacion'] > 3.5]

# Crear la gráfica de caja utilizando Seaborn
plt.figure(figsize=(12, 8))
sns.boxplot(x=df_mercado_limpio['Tasa de ocupacion'], orient='h', color='skyblue')
sns.scatterplot(x=outliers_modified_z['Tasa de ocupacion'], y=[0] * len(outliers_modified_z), color='red', label='Outliers', s=100)
plt.title('Detección de Outliers en la Tasa de Ocupación utilizando Z-Score Modificado')
plt.xlabel('Tasa de Ocupación (%)')
plt.legend()
plt.grid(True)
# Guardar el gráfico en un archivo
plt.savefig("Tasa de Ocupación utilizando Z-Score Modificado.png", bbox_inches='tight')
plt.show()

```

# Normalizacion de datos 

Vamos a analizar la correlación entre la Población en edad de trabajar y las tasas de ocupación y desocupación. Vamos a normalizar las columnas "Población en edad de trabajar", "Tasa de ocupación" y "Tasa de desocupación" y luego generar las gráficas para visualizar la correlación.

```python
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import pearsonr

# Columnas a normalizar
columnas_a_normalizar = ['Poblacion en edad de trabajar', 'Tasa de ocupacion', 'Tasa de desocupacion']

# Normalizar los datos
escalador = MinMaxScaler()
df_normalizado = df_mercado_limpio.copy()
df_normalizado[columnas_a_normalizar] = escalador.fit_transform(df_mercado_limpio[columnas_a_normalizar])

# Calcular la correlación en los datos normalizados
corr_ocupacion, _ = pearsonr(df_normalizado['Poblacion en edad de trabajar'], df_normalizado['Tasa de ocupacion'])
corr_desocupacion, _ = pearsonr(df_normalizado['Poblacion en edad de trabajar'], df_normalizado['Tasa de desocupacion'])

# Crear una figura para las gráficas de dispersión
figura, ejes = plt.subplots(1, 2, figsize=(14, 6))

# Graficar la correlación entre la población en edad de trabajar y la tasa de ocupación
sns.scatterplot(data=df_normalizado, x='Poblacion en edad de trabajar', y='Tasa de ocupacion', ax=ejes[0])
ejes[0].set_title(f'Población en Edad de Trabajar vs Tasa de Ocupación\n Correlación: {corr_ocupacion:.2f}')
ejes[0].set_xlabel('Población en Edad de Trabajar (Normalizada)')
ejes[0].set_ylabel('Tasa de Ocupación (Normalizada)')

# Graficar la correlación entre la población en edad de trabajar y la tasa de desocupación
sns.scatterplot(data=df_normalizado, x='Poblacion en edad de trabajar', y='Tasa de desocupacion', ax=ejes[1], color='r')
ejes[1].set_title(f'Población en Edad de Trabajar vs Tasa de Desocupación\n Correlación: {corr_desocupacion:.2f}')
ejes[1].set_xlabel('Población en Edad de Trabajar (Normalizada)')
ejes[1].set_ylabel('Tasa de Desocupación (Normalizada)')

# Mostrar las gráficas
plt.tight_layout()
plt.savefig("gráficos de dispersión.png", bbox_inches='tight')
plt.show()
```
